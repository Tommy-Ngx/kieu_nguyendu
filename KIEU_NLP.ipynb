```python
def main():
    try:
        doc = docx.Document('/content/Truyen-Kieu.docx')  # Creating word reader object.
        data = ""
        fullText = []
        for para in doc.paragraphs:
            fullText.append(para.text)
            data = '\n'.join(fullText)
 
        print(data)
 
    except IOError:
        print('There was an error opening the file!')
        return
 
kieutruyen = main()
kieutruyen
```
